```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1 - Variáveis e amostragem

```python
'''' Carregando o arquivo '''
dados = pd.read_csv("Arquivos/health_survey_data.csv")
print(dados.info())
```

### A -	Classifique todas as variáveis do conjunto (nominal, ordinal, discreta, contínua).

```python
variaveis_por_tipo = dados.columns.groupby(dados.dtypes)
for tipo, nomes in variaveis_por_tipo.items():
    display(f"Tipo: '{tipo}':{list(nomes)}")
```

### B - Verifique quais variáveis precisam de limpeza (valores faltantes, outliers, codificação inconsistente).

```python
dados = dados.dropna(axis=0) # Excluindo valores ausentes
display(dados)
```

```python
display(dados.describe())
```

### C - Extraia uma amostra aleatória de 100 elementos e verifique se ela é representativa da população.

```python
amostra = dados.sample(n=100) # Retornando amostra aleatória de 100 itens dos dados. 
display(amostra)
```
